## Show the GLM quality results for different cell types

```r
suppressPackageStartupMessages({
ok_aspen <- requireNamespace("ASPEN", quietly = TRUE)
if (ok_aspen) {
library(ASPEN)
} else {
message("ASPEN not installed; sourcing R/ …")
rfiles <- list.files("../R", full.names = TRUE, pattern = "*.R$")
invisible(lapply(rfiles, source))
}
library(SingleCellExperiment)
library(Matrix)
})

# Paths
sce_path <- "../data/aspensce_sexupdated_F1_filtered.rds" # change if needed
base_dir <- "../results/celltype"

# Helpers
.pick_ct_col <- function(meta) {
for (nm in c("celltype_new","celltype","celltype_old")) if (!is.null(meta[[nm]])) return(nm)
stop("No celltype column found (celltype_new/celltype/celltype_old)")
}
.parse_thresholds <- function(log_path, default_counts = 5L, default_cells = 50L) {
min_counts <- default_counts; min_cells <- default_cells
if (file.exists(log_path)) {
ln <- readLines(log_path, warn = FALSE)
m1 <- sub(".min_counts=([0-9]+).", "\1", ln[grepl("min_counts=", ln)])
m2 <- sub(".min_cells=([0-9]+).", "\1", ln[grepl("min_cells=", ln)])
if (length(m1) > 0 && grepl("^[0-9]+$", m1[1])) min_counts <- as.integer(m1[1])
if (length(m2) > 0 && grepl("^[0-9]+$", m2[1])) min_cells <- as.integer(m2[1])
}
list(min_counts = min_counts, min_cells = min_cells)
}

fit_one_gene <- function(y, n, design) {
df <- as.data.frame(design); df$resp <- y / n
fit <- tryCatch(glm(resp ~ ., family = quasibinomial(), weights = n, data = df,
control = glm.control(maxit = 100)),
error = function(e) NULL)
fit
}

# Extract per-gene coefficients for terms of interest (sex*, condition*)
coef_rows_for_terms <- function(fit, terms_regex = c("^sex", "^condition")) {
if (is.null(fit)) return(NULL)
cf <- tryCatch(summary(fit)$coefficients, error = function(e) NULL)
if (is.null(cf)) return(NULL)
rn <- rownames(cf)
out <- list()
for (rgx in terms_regex) {
idx <- grep(rgx, rn)
if (length(idx) == 0) next
for (j in idx) {
out[[length(out)+1]] <- data.frame(
term = rn[j],
estimate = cf[j,1],
se = cf[j,2],
z = cf[j,3],
p = cf[j,4],
stringsAsFactors = FALSE
)
}
}
if (length(out) == 0) return(NULL)
do.call(rbind, out)
}

analyze_celltype <- function(ct_dir, sce) {
ct <- basename(ct_dir)
est_path <- file.path(ct_dir, "estimates_global_shrunk.rds")
if (!file.exists(est_path)) { message("No estimates for ", ct); return(NULL) }
est <- readRDS(est_path)
genes <- rownames(est)

meta <- as.data.frame(colData(sce))
ct_col <- .pick_ct_col(meta)

# sex F/M labels
sex_all <- meta$sex
sex_all <- as.character(sex_all); sex_all[sex_all %in% c("Female","F")] <- "F"; sex_all[sex_all %in% c("Male","M")] <- "M"

cells_ct <- which(as.character(meta[[ct_col]]) == ct & sex_all %in% c("F","M"))
if (length(cells_ct) < 5) { message("Too few cells for ", ct); return(NULL) }

a1s <- SummarizedExperiment::assay(sce, "a1")[, cells_ct, drop = FALSE]
tots <- SummarizedExperiment::assay(sce, "tot")[, cells_ct, drop = FALSE]
genes_in <- intersect(genes, rownames(tots))
a1 <- as.matrix(a1s[genes_in, , drop = FALSE]); mode(a1) <- "integer"
tot <- as.matrix(tots[genes_in, , drop = FALSE]); mode(tot) <- "integer"

design ~ sex + condition
sex <- factor(sex_all[cells_ct], levels = c("F","M"))
cond <- meta$condition_new[cells_ct]; if (all(is.na(cond))) cond <- meta$condition_old[cells_ct]
design_df <- data.frame(sex = sex, condition = factor(ifelse(is.na(cond), "NA", as.character(cond))))
rownames(design_df) <- colnames(tot)
design <- model.matrix(~ sex + condition, data = design_df)

# Use thresholds from filter_log to match pipeline
thr <- .parse_thresholds(file.path(ct_dir, "filter_log.txt"), default_counts = 5L, default_cells = 50L)
min_counts <- thr$min_counts; min_cells <- thr$min_cells

# Fit per gene
genes_vec <- rownames(a1)
coef_list <- list()
fit_list <- list()
for (g in genes_vec) {
y <- as.numeric(a1[g, ]); n <- as.numeric(tot[g, ])
keep <- is.finite(y) & is.finite(n) & (n >= min_counts)
if (sum(keep) < min_cells) next
fit <- fit_one_gene(y[keep], n[keep], design[keep, , drop = FALSE])
if (is.null(fit)) next
# model metrics
fit_list[[g]] <- data.frame(
  gene = g,
  converged = isTRUE(fit$converged),
  iter = fit$iter,
  null_dev = fit$null.deviance,
  deviance = fit$deviance,
  df_null = fit$df.null,
  df_resid = fit$df.residual,
  r2_mcfadden = ifelse(is.finite(fit$null.deviance) && fit$null.deviance > 0, 1 - (fit$deviance / fit$null.deviance), NA_real_),
  stringsAsFactors = FALSE
)

# coefficients of interest
cf <- coef_rows_for_terms(fit, terms_regex = c("^sex", "^condition"))
if (!is.null(cf)) {
  cf$gene <- g
  coef_list[[g]] <- cf
}
}

coef_df <- if (length(coef_list) > 0) do.call(rbind, coef_list) else data.frame()
fit_df <- if (length(fit_list) > 0) do.call(rbind, fit_list) else data.frame()
list(ct = ct, coef_df = coef_df, fit_df = fit_df)
}

# Run across cell types and write outputs
sce <- readRDS(sce_path)
dirs <- list.dirs(base_dir, full.names = TRUE, recursive = FALSE)

all_coef <- list(); all_fit <- list()
for (d in dirs) {
res <- analyze_celltype(d, sce)
if (is.null(res)) next
ct <- res$ct
if (nrow(res$coef_df) > 0) {
write.csv(res$coef_df, file.path(d, "glm_coef_by_gene.csv"), row.names = FALSE)
message("Saved coef table: ", file.path(d, "glm_coef_by_gene.csv"))
}
if (nrow(res$fit_df) > 0) {
write.csv(res$fit_df, file.path(d, "glm_model_fit_by_gene.csv"), row.names = FALSE)
message("Saved model fit table: ", file.path(d, "glm_model_fit_by_gene.csv"))
}
all_coef[[ct]] <- res$coef_df
all_fit[[ct]] <- res$fit_df
}

# Aggregate quick summaries (percent significant per term; median R2)
summ_rows <- list()
for (ct in names(all_coef)) {
cf <- all_coef[[ct]]
mf <- all_fit[[ct]]
if (nrow(cf) == 0 || nrow(mf) == 0) next

# sex term(s)
sex_rows <- cf[grepl("^sex", cf$term), , drop = FALSE]
cond_rows <- cf[grepl("^condition", cf$term), , drop = FALSE]

summ_rows[[ct]] <- data.frame(
celltype = ct,
n_genes_fit = length(unique(mf$gene)),
median_r2_mcfadden = median(mf$r2_mcfadden, na.rm = TRUE),
sex_terms = paste(unique(sex_rows$term), collapse = ";"),
pct_sig_sex = mean(sex_rows$p < 0.05, na.rm = TRUE),
cond_terms = paste(unique(cond_rows$term), collapse = ";"),
pct_sig_condition = mean(cond_rows$p < 0.05, na.rm = TRUE),
stringsAsFactors = FALSE
)
}
summ_out <- if (length(summ_rows) > 0) do.call(rbind, summ_rows) else data.frame()
print(summ_out)
if (nrow(summ_out) > 0) {
write.csv(summ_out, file.path(base_dir, "glm_coef_significance_summary.csv"), row.names = FALSE)
message("Saved: ", file.path(base_dir, "glm_coef_significance_summary.csv"))
}
```

### Imbalance test

```r
suppressPackageStartupMessages({
  if (requireNamespace("ASPEN", quietly = TRUE)) {
    library(ASPEN)
  } else {
    message("ASPEN not installed; sourcing ../R …")
    rfiles <- list.files("../R", full.names = TRUE, pattern = "\\.R$")
    invisible(lapply(rfiles, source))
  }
  if (!requireNamespace("assertthat", quietly = TRUE)) install.packages("assertthat")
  library(assertthat)
  if (!requireNamespace("locfit", quietly = TRUE)) install.packages("locfit")
  library(SingleCellExperiment)
  library(Matrix)
  ok_openxlsx <- requireNamespace("openxlsx", quietly = TRUE)
  ok_readxl   <- requireNamespace("readxl", quietly = TRUE)
})

sce_path <- "../data/aspensce_sexupdated_F1_filtered.rds"
base_dir <- "../results/celltype"

.pick_ct_col <- function(meta) {
  for (nm in c("celltype_new","celltype","celltype_old"))
    if (!is.null(meta[[nm]])) return(nm)
  stop("No celltype column found (celltype_new/celltype/celltype_old)")
}

load_exclusions <- function() {
  genes_excl <- character(0)
  xy_path <- system.file("extdata", "mm10_genesXY.txt", package = "ASPEN")
  if (nzchar(xy_path) && file.exists(xy_path)) {
    genesXY <- tryCatch(read.table(xy_path), error = function(e) NULL)
    if (!is.null(genesXY)) genes_excl <- c(genes_excl, as.character(genesXY[[1]]))
  }
  impr_path <- system.file("extdata", "mm10_imprinted_genes.xlsx", package = "ASPEN")
  if (nzchar(impr_path) && file.exists(impr_path)) {
    genesIMPR <- NULL
    if (ok_openxlsx) {
      genesIMPR <- tryCatch(openxlsx::read.xlsx(impr_path, colNames = TRUE), error = function(e) NULL)
    } else if (ok_readxl) {
      genesIMPR <- tryCatch(as.data.frame(readxl::read_excel(impr_path)), error = function(e) NULL)
    }
    if (!is.null(genesIMPR)) {
      col <- intersect(c("imprinted.genes","gene","symbol"), colnames(genesIMPR))
      if (length(col) > 0) genes_excl <- c(genes_excl, as.character(genesIMPR[[col[1]]]))
    }
  }
  unique(genes_excl)
}

clean_params <- function(x) {
  nm <- names(x)
  nm <- sub("\\..*$", "", nm)    # drop trailing “.1” etc.
  names(x) <- nm
  x
}

run_glob_disp_ct <- function(a1, tot, genes_excl,
                             min_counts_candidates = c(5, 2, 1),
                             max_pairs = 2e6,
                             verbose = TRUE) {
  stopifnot(identical(dim(a1), dim(tot)))
  mode(a1) <- "integer"; mode(tot) <- "integer"
  rn <- rownames(tot); if (is.null(rn)) rn <- paste0("gene_", seq_len(nrow(tot)))

  lbetabin <- function(df, inits){
    y <- df$y; n <- df$n
    alpha <- inits[1]; beta <- inits[2]
    sum(lchoose(n, y) + lgamma(alpha + beta) - lgamma(n + alpha + beta) +
          lgamma(y + alpha) - lgamma(alpha) + lgamma(n - y + beta) - lgamma(beta))
  }

  for (mc in min_counts_candidates) {
    if (verbose) cat("\nTrying min_counts =", mc, "\n")
    idx <- which(tot > mc, arr.ind = TRUE)
    if (!nrow(idx)) { if (verbose) cat("  No pairs at this threshold\n"); next }

    gene_names <- rn[idx[,1]]
    keep <- !(gene_names %in% genes_excl)
    idx <- idx[keep, , drop = FALSE]
    if (!nrow(idx)) { if (verbose) cat("  All excluded at this threshold\n"); next }

    if (nrow(idx) > max_pairs) {
      if (verbose) cat("  Downsampling pairs from", nrow(idx), "to", max_pairs, "\n")
      set.seed(1)
      idx <- idx[sample.int(nrow(idx), max_pairs), , drop = FALSE]
    }

    y <- a1[idx]; n <- tot[idx]
    df <- data.frame(y = as.numeric(y), n = as.numeric(n))
    df <- df[is.finite(df$y) & is.finite(df$n) & df$n > mc, , drop = FALSE]
    if (nrow(df) < 1000) { if (verbose) cat("  Too few rows after filter:", nrow(df), "\n"); next }

    binfit <- tryCatch(glm((y/n) ~ 1, family = binomial(), weights = n, data = df),
                       error = function(e) NULL)
    if (is.null(binfit) || !is.finite(binfit$fitted.values[1])) {
      if (verbose) cat("  Binomial init failed; using mean(y/n)\n")
      p_init <- mean(df$y / df$n)
    } else {
      p_init <- binfit$fitted.values[1]
    }
    p_init <- min(max(p_init, 1e-3), 1 - 1e-3)
    inits <- c(p_init, 1 - p_init)

    opt1 <- tryCatch(optim(inits, lbetabin, df = df, method = "L-BFGS-B",
                           lower = c(1e-2, 1e-2), upper = c(1e6, 1e6),
                           control = list(fnscale = -1)), error = function(e) NULL)
    if (!is.null(opt1) && is.finite(opt1$value) && !is.null(opt1$par)) {
      alpha <- opt1$par[1]; beta <- opt1$par[2]
    } else {
      if (verbose) cat("  L-BFGS-B failed; trying Nelder–Mead\n")
      opt2 <- tryCatch(optim(inits, lbetabin, df = df, method = "Nelder-Mead",
                             control = list(fnscale = -1, maxit = 2000)),
                       error = function(e) NULL)
      if (!is.null(opt2) && is.finite(opt2$value) && !is.null(opt2$par)) {
        alpha <- opt2$par[1]; beta <- opt2$par[2]
      } else {
        if (verbose) cat("  All MLE attempts failed; using method-of-moments fallback\n")
        p_bar <- mean(df$y / df$n)
        var_y <- var(df$y / df$n)
        m_eff <- sum(df$n * p_bar * (1 - p_bar)) / sum(p_bar * (1 - p_bar))
        rho <- (var_y - (p_bar * (1 - p_bar) / mean(df$n))) / (p_bar * (1 - p_bar))
        rho <- max(0, min(rho, 0.999))
        theta <- rho / (1 - rho)
        alpha <- p_bar / max(theta, 1e-8)
        beta  <- (1 - p_bar) / max(theta, 1e-8)
      }
    }

    mu <- alpha / (alpha + beta)
    theta <- 1 / (alpha + beta)
    if (verbose) cat(sprintf("  Success: mu=%.4f theta=%.4g (alpha=%.3f, beta=%.3f)\n",
                             mu, theta, alpha, beta))
    return(list(params = c(mu = mu, theta = theta, alpha = alpha, beta = beta),
                used_min_counts = mc))
  }
  if (verbose) cat("  No success at any min_counts\n")
  NULL
}

# run per cell type
sce  <- readRDS(sce_path)
meta <- as.data.frame(colData(sce))
ct_col <- .pick_ct_col(meta)

sex_all <- meta$sex
sex_all <- as.character(sex_all)
sex_all[sex_all %in% c("Female","F")] <- "F"
sex_all[sex_all %in% c("Male","M")]   <- "M"
cells_keep <- which(sex_all %in% c("F","M"))

cells_by_ct <- split(colnames(sce)[cells_keep], meta[cells_keep, ct_col])
genes2remove <- load_exclusions()

glob_list <- list()
for (ct in names(cells_by_ct)) {
  cols <- cells_by_ct[[ct]]
  if (length(cols) < 5) { message("Skip ", ct, " (too few cells)"); next }

  a1  <- as.matrix(assays(sce)[["a1"]][, cols, drop = FALSE]); mode(a1)  <- "integer"
  tot <- as.matrix(assays(sce)[["tot"]][, cols, drop = FALSE]); mode(tot) <- "integer"

  res <- run_glob_disp_ct(a1, tot, genes2remove, min_counts_candidates = c(5, 2, 1), max_pairs = 2e6)
  if (is.null(res)) { message("glob_disp failed for ", ct, " (even at min_counts = 1)"); next }

  gp <- clean_params(res$params)   # names: mu, theta, alpha, beta
  used_mc <- res$used_min_counts
  glob_list[[ct]] <- gp

  message(sprintf("Cell type %s: mu=%.4f, theta=%.4g, (alpha=%.3f, beta=%.3f), min_counts used=%d",
                  ct, gp["mu"], gp["theta"], gp["alpha"], gp["beta"], used_mc))

  out_dir <- file.path(base_dir, ct)
  if (dir.exists(out_dir)) {
    out_path <- file.path(out_dir, "global_params.csv")
    df <- data.frame(mu = gp["mu"], theta = gp["theta"],
                     alpha = gp["alpha"], beta = gp["beta"],
                     min_counts_used = used_mc,
                     row.names = ct)
    write.table(df, out_path, quote = FALSE, sep = ",",
                col.names = !file.exists(out_path), append = file.exists(out_path))
    message("Saved: ", out_path)
  }
}
glob_list
```

```r
use_norm <- TRUE  # FALSE = raw-input test, TRUE = normalized-input test

sce  <- readRDS("../data/aspensce_sexupdated_F1_filtered.rds")
meta <- as.data.frame(colData(sce))
ct_col <- intersect(c("celltype_new","celltype","celltype_old"), names(meta))[1]

sex_all <- as.character(meta$sex)
sex_all[sex_all %in% c("Female","F")] <- "F"
sex_all[sex_all %in% c("Male","M")]   <- "M"

cells_keep <- which(sex_all %in% c("F","M"))
cells_by_ct <- split(colnames(sce)[cells_keep], meta[cells_keep, ct_col])

min_counts   <- 5   # same threshold the pipeline used for glob_disp and tests
min_cells    <- 10
genes_excl   <- load_exclusions()

for (ct in names(cells_by_ct)) {
  cols <- cells_by_ct[[ct]]
  if (length(cols) < 5) next

  est_path <- file.path("../results/celltype", ct, "estimates_global_shrunk.rds")
  if (!file.exists(est_path)) { message("No estimates for ", ct); next }
  est_shrunk <- readRDS(est_path)

  # recover the raw counts for this CT
  a1_raw  <- as.matrix(assays(sce)[["a1"]][, cols, drop = FALSE]);  mode(a1_raw)  <- "integer"
  tot_raw <- as.matrix(assays(sce)[["tot"]][, cols, drop = FALSE]); mode(tot_raw) <- "integer"

  # align genes
  common <- intersect(rownames(est_shrunk), rownames(tot_raw))
  if (!length(common)) next
  est_ct  <- est_shrunk[common, , drop = FALSE]
  a1_ct   <- a1_raw[common, , drop = FALSE]
  tot_ct  <- tot_raw[common, , drop = FALSE]

  # normalized view if desired
  if (use_norm) {
    sf <- Matrix::colSums(tot_raw[, cols, drop = FALSE])
    sf <- sf / exp(mean(log(sf)))
    tot_ct <- sweep(tot_ct, 2, sf, "/")
    a1_ct  <- sweep(a1_ct,  2, sf, "/")
  }

  # global μ stays from raw counts
  gp_path <- file.path("../results/celltype", ct, "global_params.csv")
  if (!file.exists(gp_path)) { message("No global params for ", ct); next }
  gp_df <- read.csv(gp_path, row.names = 1, check.names = FALSE)
  gp <- unlist(gp_df[ct, c("mu","theta","alpha","beta")])
  names(gp) <- c("mu","theta","alpha","beta")

  res_bb <- bb_mean(a1_counts = a1_ct,
                    tot_counts = tot_ct,
                    estimates  = est_ct,
                    glob_params = gp,
                    min_cells = min_cells,
                    min_counts = min_counts)

  suffix <- if (use_norm) "norm" else "rawrep"
  out_dir <- file.path("../results/celltype", ct)
  write.csv(as.data.frame(res_bb), file.path(out_dir, sprintf("bb_mean_results_%s.csv", suffix)), row.names = TRUE)
  saveRDS(res_bb, file.path(out_dir, sprintf("bb_mean_results_%s.rds", suffix)))
}
```

```r
## (3) Significant sex-effect genes (Cardiomyocyte & Fibroblast)
sig_sex_genes <- function(ct, alpha = 0.05) {
    coef_path <- file.path("../results", "celltype", ct, "glm_coef_by_gene.csv")
    if (!file.exists(coef_path)) {
        message("Missing coefficient file for ", ct);
        return(list(genes = character(), table = data.frame()))
    }
    coef_tbl <- read.csv(coef_path, stringsAsFactors = FALSE)
    if (!nrow(coef_tbl)) return(list(genes = character(), table = coef_tbl[0, ]))
    sex_rows <- subset(coef_tbl, grepl('^sex', term) & is.finite(p))
    sig_rows <- subset(sex_rows, p < alpha)
    list(
        genes = unique(sig_rows$gene),
        table = sig_rows[order(sig_rows$p), ]
    )
}

alpha_level <- 0.05
cts_to_check <- c("Cardiomyocyte", "Fibroblast")

sex_hits <- lapply(cts_to_check, sig_sex_genes, alpha = alpha_level)
names(sex_hits) <- cts_to_check

signif_counts <- vapply(sex_hits, function(x) length(x$genes), integer(1))
print(signif_counts)

sex_hits_df <- lapply(names(sex_hits), function(ct) {
    df <- sex_hits[[ct]]$table
    if (!nrow(df)) return(NULL)
    df$celltype <- ct
    df[, c("celltype", "gene", "term", "estimate", "se", "z", "p")]
})
sex_hits_df <- do.call(rbind, sex_hits_df)

if (!is.null(sex_hits_df) && nrow(sex_hits_df)) {
    print(head(sex_hits_df, 20))
    write.csv(sex_hits_df,
              file = file.path("results", "sex_effect_significant_genes_cardio_fibro.csv"),
              row.names = FALSE)
    message("Saved: results/sex_effect_significant_genes_cardio_fibro.csv")
} else {
    message("No significant sex-effect genes found for the selected cell types at alpha = ", alpha_level)
}
```

```r
## (4) GLM coefficient utilities (shared helpers)
load_glm_coef_table <- function(ct) {
    candidates <- c(
        file.path("results", "celltype", ct, "glm_coef_by_gene.csv"),
        file.path("..", "results", "celltype", ct, "glm_coef_by_gene.csv")
    )
    idx <- which(file.exists(candidates))[1]
    if (is.na(idx)) {
        stop(sprintf("Coefficient table not found for %s (searched: %s)", ct, paste(candidates, collapse = ", ")))
    }
    read.csv(candidates[idx], stringsAsFactors = FALSE)
}

extract_sig_terms <- function(ct, pattern = "^sex", alpha = 0.05) {
    tbl <- tryCatch(load_glm_coef_table(ct), error = function(e) NULL)
    if (is.null(tbl) || !nrow(tbl)) {
        message("No coefficients available for ", ct)
        return(list(genes = character(), table = tbl[0, ]))
    }
    keep <- grepl(pattern, tbl$term) & is.finite(tbl$p)
    sub_tbl <- tbl[keep, , drop = FALSE]
    sig_tbl <- sub_tbl[sub_tbl$p < alpha, , drop = FALSE]
    sig_tbl <- sig_tbl[order(sig_tbl$p), , drop = FALSE]
    list(
        genes = unique(sig_tbl$gene),
        table = sig_tbl
    )
}

build_signed_logp_matrix <- function(ct, terms_regex = "^(sex|condition)", top_genes = 40) {
    tbl <- tryCatch(load_glm_coef_table(ct), error = function(e) NULL)
    if (is.null(tbl) || !nrow(tbl)) {
        message("No coefficients available for ", ct)
        return(list(matrix = NULL, table = tbl[0, ]))
    }
    keep <- grepl(terms_regex, tbl$term) & is.finite(tbl$p)
    tbl <- tbl[keep, , drop = FALSE]
    if (!nrow(tbl)) {
        message("No matching coefficient terms for ", ct)
        return(list(matrix = NULL, table = tbl))
    }
    tbl$score <- sign(tbl$estimate) * -log10(pmax(tbl$p, .Machine$double.xmin))
    tbl <- tbl[order(tbl$p), , drop = FALSE]
    genes <- unique(tbl$gene)
    if (length(genes) > top_genes) genes <- genes[seq_len(top_genes)]
    tbl_top <- tbl[tbl$gene %in% genes, , drop = FALSE]
    terms <- sort(unique(tbl_top$term))
    mat <- matrix(NA_real_, length(genes), length(terms),
                  dimnames = list(genes, terms))
    if (nrow(tbl_top)) {
        for (i in seq_len(nrow(tbl_top))) {
            g <- tbl_top$gene[i]
            term <- tbl_top$term[i]
            mat[g, term] <- tbl_top$score[i]
        }
    }
    list(matrix = mat, table = tbl_top)
}
```

```r
## Two per-term heatmaps of signed -log10(p): one for SEX, one for CONDITION
suppressPackageStartupMessages({
  if (!"ComplexHeatmap" %in% loadedNamespaces()) library(ComplexHeatmap)
  if (!"circlize"        %in% loadedNamespaces()) library(circlize)
})

# --- helpers -------------------------------------------------------------
.standardize_term <- function(x) {
  x <- sub("^sexM$",               "sex",       x)
  x <- sub("^sex$",                "sex",       x)
  x <- sub("^conditionF1_Young$",  "condition", x)
  x <- sub("^F1_Young$",           "condition", x)
  x <- sub("^F1young$",            "condition", x, ignore.case = TRUE)
  x <- sub("^condition$",          "condition", x)
  x
}

.collapse_dup_cols <- function(mat) {
  if (!length(colnames(mat)) || !any(duplicated(colnames(mat)))) return(mat)
  uniq <- unique(colnames(mat))
  out  <- sapply(uniq, function(nm) {
    jj <- which(colnames(mat) == nm)
    if (length(jj) == 1L) mat[, jj]
    else rowMeans(mat[, jj, drop = FALSE], na.rm = TRUE)
  })
  out <- as.matrix(out); rownames(out) <- rownames(mat); colnames(out) <- uniq
  out
}

.pick_top_genes_from_table <- function(tab, term_name, n = 40, exclude = character()) {
  stopifnot(all(c("term","gene","p","z","estimate") %in% names(tab)))
  tt <- tab
  tt$term <- .standardize_term(tt$term)
  tt$gene <- as.character(tt$gene)
  tt <- subset(tt, term == term_name & is.finite(p))
  if (!nrow(tt)) return(character())
  ord <- with(tt, order(p, -abs(z), -abs(estimate)))
  genes <- setdiff(tt$gene[ord], exclude)
  head(unique(genes), n)
}

.cap_and_limits <- function(x, quantile_cap = 0.95, min_endcap = 1) {
  v <- as.numeric(x)
  pos <- v[v > 0 & is.finite(v)]
  neg <- -v[v < 0 & is.finite(v)]
  pos_lim <- if (length(pos)) max(min_endcap, as.numeric(stats::quantile(pos, quantile_cap, na.rm = TRUE))) else min_endcap
  neg_lim <- if (length(neg)) max(min_endcap, as.numeric(stats::quantile(neg, quantile_cap, na.rm = TRUE))) else min_endcap
  list(pos = pos_lim, neg = -neg_lim,
       cap = function(m) pmax(pmin(m, pos_lim), -neg_lim))
}

.make_single_term_heatmap <- function(ct_label, term_label, submat, lims) {
  col_fun <- circlize::colorRamp2(c(lims$neg, 0, lims$pos), c("blue","white","red"))
  ht <- Heatmap(submat,
                name = paste0("signed_log10p_", term_label),
                col = col_fun, na_col = "grey90",
                cluster_rows = TRUE, cluster_columns = FALSE,
                show_row_names = TRUE, show_column_names = TRUE,
                show_heatmap_legend = FALSE,
                column_title = paste(ct_label, "—", term_label))
  lgd_neg <- Legend(
    title = expression(-log[10]*"(p) ("*"\u2212"*")"),
    col_fun = circlize::colorRamp2(c(lims$neg, 0), c("blue","white")),
    at = pretty(c(lims$neg, 0), n = 3)
  )
  lgd_pos <- Legend(
    title = expression(-log[10]*"(p) ("*"+"*")"),
    col_fun = circlize::colorRamp2(c(0, lims$pos), c("white","red")),
    at = pretty(c(0, lims$pos), n = 3)
  )
  list(ht = ht, legends = list(lgd_neg, lgd_pos))
}

# General saver (pdf/png/tiff/jpeg)
.save_ht <- function(htl, out_file, height, width = 5.5, dpi = 150) {
  dir.create(dirname(out_file), recursive = TRUE, showWarnings = FALSE)
  ext <- tolower(tools::file_ext(out_file))
  dev_id <- NULL
  if (ext == "pdf") {
    grDevices::pdf(out_file, width = width, height = height, useDingbats = FALSE)
    dev_id <- grDevices::dev.cur()
  } else if (ext %in% c("png","jpg","jpeg","tif","tiff")) {
    fun <- switch(ext,
                  png  = grDevices::png,
                  jpg  = grDevices::jpeg,
                  jpeg = grDevices::jpeg,
                  tif  = grDevices::tiff,
                  tiff = grDevices::tiff)
    fun(out_file, width = width, height = height, units = "in", res = dpi)
    dev_id <- grDevices::dev.cur()
  } else stop("Unsupported extension: ", ext)
  on.exit(try(grDevices::dev.off(dev_id), silent = TRUE), add = TRUE)
  draw(htl$ht, newpage = TRUE, heatmap_legend_list = htl$legends, heatmap_legend_side = "bottom")
  invisible(out_file)
}

# Backward-compatible PDF specific helper (calls generic)
.save_ht_to_pdf <- function(htl, out_pdf, height, width = 5.5) {
  .save_ht(htl, out_pdf, height = height, width = width)
}

# --- main ----------------------------------------------------------------
plot_two_term_heatmaps <- function(
  ct,
  terms_regex   = "^(sex|sexM|condition|F1young|F1_Young|conditionF1_Young)$",
  top_n         = 40,
  disjoint_sets = TRUE,
  show_plot     = TRUE,
  quantile_cap  = 0.95,
  min_endcap    = 1,
  extra_formats = c("png")   # add other wanted formats, e.g. c("png","tiff")
) {
  res <- build_signed_logp_matrix(ct, terms_regex = terms_regex, top_genes = Inf)
  mat <- res$matrix
  if (is.null(mat) || !is.matrix(mat) || nrow(mat) == 0 || all(is.na(mat))) {
    message("Skipping ", ct, " (no coefficient data).")
    return(invisible(NULL))
  }

  colnames(mat) <- .standardize_term(colnames(mat))
  mat <- .collapse_dup_cols(mat)
  keep <- intersect(c("sex", "condition"), colnames(mat))
  if (!length(keep)) {
    message("No 'sex' or 'condition' columns for ", ct, ".")
    return(invisible(NULL))
  }
  mat <- mat[, keep, drop = FALSE]

  sex_genes  <- if ("sex" %in% keep)       .pick_top_genes_from_table(res$table, "sex",       n = top_n) else character()
  cond_excl  <- if (disjoint_sets) sex_genes else character()
  cond_genes <- if ("condition" %in% keep)  .pick_top_genes_from_table(res$table, "condition", n = top_n, exclude = cond_excl) else character()

  outs <- list()

  if (length(sex_genes)) {
    g <- sex_genes[sex_genes %in% rownames(mat)]
    sex_mat <- mat[g, "sex", drop = FALSE]
    lims    <- .cap_and_limits(sex_mat[,1], quantile_cap, min_endcap)
    sex_mat <- lims$cap(sex_mat)
    htl     <- .make_single_term_heatmap(ct, "sex", sex_mat, lims)

    base_dir <- file.path("results", "celltype", ct)
    sex_pdf  <- file.path(base_dir, sprintf("glm_coef_signed_logp_heatmap_%s_sex.pdf", ct))
    pdf_h    <- max(4, min(14, nrow(sex_mat) * 0.22 + 2))
    .save_ht_to_pdf(htl, sex_pdf, height = pdf_h)

    # extra formats
    extra_paths <- vapply(extra_formats, function(ext) {
      f <- sub("\\.pdf$", paste0(".", ext), sex_pdf)
      .save_ht(htl, f, height = pdf_h)
    }, character(1))

    # save Heatmap object as RDS
    sex_rds <- sub("\\.pdf$", "_object.rds", sex_pdf)
    saveRDS(htl$ht, sex_rds)

    if (isTRUE(show_plot)) {
      draw(htl$ht, newpage = TRUE, heatmap_legend_list = htl$legends, heatmap_legend_side = "bottom")
    }

    outs$sex <- list(genes = g, pdf = sex_pdf, extra = extra_paths, rds = sex_rds, limits = lims)
  } else message("No 'sex' genes selected for ", ct, ".")

  if (length(cond_genes)) {
    g <- cond_genes[cond_genes %in% rownames(mat)]
    cond_mat <- mat[g, "condition", drop = FALSE]
    lims     <- .cap_and_limits(cond_mat[,1], quantile_cap, min_endcap)
    cond_mat <- lims$cap(cond_mat)
    htl      <- .make_single_term_heatmap(ct, "condition", cond_mat, lims)

    base_dir <- file.path("results", "celltype", ct)
    cond_pdf <- file.path(base_dir, sprintf("glm_coef_signed_logp_heatmap_%s_condition.pdf", ct))
    pdf_h    <- max(4, min(14, nrow(cond_mat) * 0.22 + 2))
    .save_ht_to_pdf(htl, cond_pdf, height = pdf_h)

    extra_paths <- vapply(extra_formats, function(ext) {
      f <- sub("\\.pdf$", paste0(".", ext), cond_pdf)
      .save_ht(htl, f, height = pdf_h)
    }, character(1))

    cond_rds <- sub("\\.pdf$", "_object.rds", cond_pdf)
    saveRDS(htl$ht, cond_rds)

    if (isTRUE(show_plot)) {
      draw(htl$ht, newpage = TRUE, heatmap_legend_list = htl$legends, heatmap_legend_side = "bottom")
    }

    outs$condition <- list(genes = g, pdf = cond_pdf, extra = extra_paths, rds = cond_rds, limits = lims)
  } else message("No 'condition' genes selected for ", ct, ".")

  invisible(outs)
}

## Run it
coef_heatmap_outputs <- list(
  Cardiomyocyte = plot_two_term_heatmaps("Cardiomyocyte", top_n = 40, extra_formats = c("png","tiff")),
  Fibroblast    = plot_two_term_heatmaps("Fibroblast",    top_n = 40, extra_formats = c("png","tiff"))
)
```
